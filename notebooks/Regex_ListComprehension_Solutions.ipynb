## Exercise 1 – Extract Only Words

```python
import re

text = "Dr. Rossi arrived at 10:30 a.m. on 12/05/2024!"
words = re.findall(r"[A-Za-z]+", text)
words
```

## Exercise 2 – Extract Numbers

```python
text = "The patient has 3 symptoms and 2 previous surgeries."
nums = [int(x) for x in re.findall(r"\d+", text)]
nums
```

## Exercise 3 – Keep Apostrophes in Words

```python
text = "L'Aprente è l'uomo. Non è semplice!"
tokens = re.findall(r"[A-Za-zÀ-ÖØ-öø-ÿ]+(?:'[A-Za-zÀ-ÖØ-öø-ÿ]+)?", text)
tokens
```

## Exercise 4 – Extract Emails

```python
text = "Contact us at info@hospital.it or support@mail.com"
emails = re.findall(r"[\w\.-]+@[\w\.-]+\.\w+", text)
emails
```

## Exercise 7 – Lowercase and Remove Short Words

```python
tokens = ["Alice", "in", "Wonderland", "is", "fun"]
clean = [t.lower() for t in tokens if len(t) >= 3]
clean
```

## Exercise 8 – Word Length Tuples

```python
tokens = ["alice", "rabbit", "tea", "hat"]
pairs = [(t, len(t)) for t in tokens]
pairs
```

## Exercise 9 – Remove Stopwords

```python
tokens = ["the", "rabbit", "runs", "in", "the", "garden"]
stopwords = {"the", "in"}
filtered = [t for t in tokens if t not in stopwords]
filtered
```

## Exercise 10 – Tokenizer with Regex and Stopwords

```python
text = "Alice, running quickly, chased the rabbit."
stopwords = {"the"}

tokens = re.findall(r"[A-Za-z]+", text)
tokens = [t.lower() for t in tokens if t.lower() not in stopwords]
tokens
```

## Exercise 11 – Vocabulary Size

```python
docs = [
    "Alice loves the rabbit",
    "The rabbit loves carrots"
]
stopwords = {"the"}

tokenized_docs = [
    [t.lower() for t in re.findall(r"[A-Za-z]+", doc) if t.lower() not in stopwords]
    for doc in docs
]

vocab = set(t for doc in tokenized_docs for t in doc)
len(vocab), sorted(vocab)
```

## Exercise 13 – Consecutive Repeated Words

```python
text = "Alice Alice rabbit rabbit rabbit hat"

pattern = re.compile(r"\b(\w+)(?:\s+\1)+\b", flags=re.IGNORECASE)
full_matches = [m.group(0) for m in pattern.finditer(text)]
full_matches
```

## Exercise 14 – Extract Bigrams

```python
tokens = ["alice", "in", "wonderland"]
bigrams = [f"{tokens[i]} {tokens[i+1]}" for i in range(len(tokens)-1)]
bigrams
```
